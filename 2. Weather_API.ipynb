```python
temp_min = []
temp_max = []
feels_like = []
day = []

# Combine the different parts from above to create one url.
response = requests.get(f"https://api.openweathermap.org/data/2.5/forecast?lat={latitude}&lon={longitude}&appid={API_key}&units=metric")
wth_json = response.json()
wth_list = wth_json ["list"]
#wth_list


for i in range(len(wth_list)):
    temp_min.append(wth_list[i]["main"]["temp_min"])
    temp_max.append(wth_list[i]["main"]["temp_max"])
    feels_like.append(wth_list[i]["main"]["feels_like"])
    day.append(wth_list[i]["dt_txt"])
    
```
